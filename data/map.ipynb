```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df1 = pd.read_csv('dataset1_info.csv')

# 1. Create the base map with the original data
fig = px.scatter_geo(df1,
                     lat='latitude',
                     lon='longitude',
                     title='Geographical Distribution with Highlighted Point')

# 2. Add the specific point as a new, styled trace
fig.add_trace(go.Scattergeo(
    lat=[48.85342445915592],
    lon=[2.3539870977401733],
    mode='markers',
    marker=dict(
        color='red',
        size=15
    ),
    name='Paris (Highlighted)'
))

# 3. Apply the custom box layout
fig.update_geos(
    oceancolor="white",
    showocean=True,
    landcolor="gainsboro",
    showland=True,
    showframe=True,
    framecolor="black"
)

fig.show()
```

```python
# 1. Load data from dataset2
df2 = pd.read_csv('dataset2_info.csv')

# 2. Create the base map with a custom width and height
fig = px.scatter_geo(df2,
                     lat='latitude',
                     lon='longitude',
                     hover_name='city',
                     scope='north america',
                     title='Data Distribution in the United States',
                     width=1200,
                     height=600)

# 3. Add the new highlighted point
fig.add_trace(go.Scattergeo(
    lat=[20.8261235],
    lon=[-156.9181519],
    mode='markers',
    marker=dict(
        color='red',
        size=15
    ),
    name='Highlighted Point'
))

# 4. Apply the custom box layout
fig.update_geos(
    oceancolor="white",
    showocean=True,
    landcolor="gainsboro",
    showland=True,
    showframe=True,
    framecolor="black"
)

fig.show()
```
